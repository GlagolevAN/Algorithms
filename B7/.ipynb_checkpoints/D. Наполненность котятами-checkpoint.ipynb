```python
def fullfillness(cats, intervals, n, m):
    arr = [0] * (n + 2 * m)
    for i in range(n):
        arr[i] = (cats[i], 0)
    for i in range(m):
        arr[n + 2 * i], arr[n + 2 * i + 1] = (intervals[2 * i], -1, i), (intervals[2 * i + 1], 1, i)
    arr.sort()
    
    cats_on_inter = [0] * m
    cat_counter = 0
    
    for elem in arr:
        if elem[1] == -1:
            cats_on_inter[elem[2]] = cat_counter
        elif elem[1] == 1:
            cats_on_inter[elem[2]] = cat_counter - cats_on_inter[elem[2]]
        else:
            cat_counter += 1
            
    return cats_on_inter
        
with open('input.txt', 'r') as fin:
    n, m = list(map(int, fin.readline().split()))
    
    cats = list(map(int, fin.readline().split()))
    
    intervals = [0] * 2*m
    for i in range(m):
        intervals[2 * i], intervals[2 * i + 1] = list(map(int, fin.readline().split()))
        
    print(*fullfillness(cats, intervals, n, m))
```
